Listas sin orden, que contiene elementos sin repetición

```python
conjunto = {4, 2, 3, 9,'hello'}
print(conjunto)
```

Si repetimos, no pasa nada, internamente sólo se guarda uno.

```python
conjunto = {4, 2, 2, 2, 2, 2, 3, 9, 'hello'}
print(conjunto)


for elemento in conjunto:
    print(elemento)
    
a=2
if a in conjunto:
    print(f"El número {a} está en el conjunto")
else:
    print(f"El número {a} no está en el conjunto")
    
#eliminar un elemento del conjunto
conjunto.remove(2)
print(conjunto)

conjunto.remove(7)  # devuelve error porque no está en el conjunto
print(conjunto)

a=7
if a in conjunto:
    conjunto.remove(7)
    
# añadir un lemento al conjunto
conjunto.add(7)
print(conjunto)
    
```

# 3. Funciones
Una función es un bloque de codigo que recrea una cierta acción. 

Definición de una función

```python
def am_i_happy(happiness_level):
    if happiness_level >= 10:
        return "You're very happy."
    elif happiness_level >= 5:
        return "You're happy."
    else:
        return "You're not happy."
```

Llamada a la función

```python
print(am_i_happy(0))
```

```python
print(am_i_happy(5))
```

# 4. Control de Flujo
## 4.1 If/Else If/Else

```python
sleepy = True
hungry = False

if sleepy and hungry:
    print("Eat a snack and take a nap.")
elif sleepy and not hungry:
    print("Take a nap")
elif hungry and not sleepy:
    print("Eat a snack")
else:
    print("Go on with your day")
```

# Clases y herencia

*\_\_init\_\_()* es el método constructor. Las clases hijo heredan todo del padre.

*super()* llama al constructor de la clase padre.





```python
# Creamos la clase Animal
class Animal:
  def __init__(self, name):
    self.name = name
    self.sound = None
  
  def emit_sound(self):
    print(f"[{type(self).__name__}] {self.name}: {self.sound}")
```

```python
#Creamos las clases Perro Gato y Cerdo que heredan de Animal
class Dog(Animal):
  def __init__(self, name):
    super().__init__(name)
    self.sound = "bark!"
    
class Cat(Animal):
  def __init__(self, name):
    super().__init__(name)
    self.sound = "meow!"

class Pig(Animal):
  def __init__(self, name):
    super().__init__(name)
    self.sound = "oink!"
```

```python
#Creamos instancias de los objetos perro gato y cerdo 
dog = Dog("Alice")
dog.emit_sound() #lanzamos el metodo "emitir_sonido" del objeto

cat = Cat("Bob")
cat.emit_sound()#lanzamos el metodo "emitir_sonido" del objeto


pig = Pig("Charly")
pig.emit_sound()#lanzamos el metodo "emitir_sonido" del objeto
```

[Métodos especiales](https://docs.python.org/3/reference/datamodel.html#specialnames) como *\_\_str\_\_()* se pueden usar para sobrecargar funciones nativas:

```python
print(dog)
```

```python
class BetterDog(Dog):
  def __init__(self, name):
    super().__init__(name)
  
  def __str__(self):
    return f"[{type(self).__name__}] {self.name}"
```

```python
smart_dog = BetterDog("Tom")
print(smart_dog)
```

# Comentario sobre formas de importar un módulo

Pongamos la función `sqrt` del módulo `math`. Se puede importar el módulo ´math´ al completo, lo que nos proporciona un objeto representando todos los atributos y métodos del módulo, y llamar al método ´sqrt´



```python
import math
math.sqrt(2)
```

**Ahora reiniciamos el entorno de ejecución para deshacer el efecto de la importación anterior.**

La forma anterior importa *todo* el módulo al completo. También podemos importar sólo un elemento de un módulo de la siguiente manera:

```python
from math import sqrt
math.sqrt(2)
```

Sin embargo ahora no podemos llamar a la función de la misma forma. Cuando importamos sólo parte de un módulo, se importa como si partiera del módulo actual, es decir:

```python
sqrt(2)
```

```python
# También podemos importar una selección de funciones
from math import log, log10, log2
```

```python
def sqrt(x):
  # Supongamos que mi versión custom de sqrt tiene un comportamiento distinto al habitual
  return "Ups!" if x <= 0 else x ** 0.5

sqrt(-3)
```

```python
from math import sqrt as math_sqrt
print(f"Mi sqrt(0) es {sqrt(0)} y math.sqrt(-1) es {math_sqrt(0)}")
```

**Volvemos a reiniciar el entorno de ejecución.**

Finalmente, si queremos importar muchas funciones pero no queremos tener que escribir el nombre del módulo siempre que hagamos uso de ellas, podemos importar todo el módulo 

```python
from math import *
print(f"Tengo acceso a log(10)={log(10):.2f}, a abs(-10)={abs(-10)} y todo lo que yo quiera de math")
```

# Excepciones

Python sigue la filosofía de 'mejor pedir perdón que pedir permiso'. Se recomienda catchear posibles excepciones y luego controlarlas más que asegurar que el input es correcto antes de ejecutar código problemático.

```python
num = 15
for i in range(10):
  try:
    modulus = num % i
  except ZeroDivisionError:
    modulus = None
  print(f"{num} módulo {i} es igual a {modulus}")
```

```python
# De forma más general

num = 15
for i in range(10):
  try:
    modulus = num % i
  except Exception as e:
    print(f"Se ha tenido una exepción para i={i}: \n\t{e}")
    modulus = None
  print(f"{num} módulo {i} es igual a {modulus}")
```
