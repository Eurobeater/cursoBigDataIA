### Práctiva 17 Clustering básico con K-Means

En esta práctica vamos a entrenar un modelo K-Means y vamos a trabajar con sus metricas básicas


##### ARI (Adjusted Rand Index)

El ARI (Adjusted Rand Index) se traduce como Índice de Rand Ajustado. Es una métrica que mide la similitud entre dos agrupaciones (clusters), comparando qué tan bien se alinean los grupos predichos por el algoritmo con las etiquetas reales de los datos.

Tiene un rango de -1 a 1, donde:

1: Coincidencia perfecta entre los clusters y las etiquetas reales.

0: Coincidencia aleatoria (los clusters no tienen relación con las etiquetas reales).

Valores negativos: Peor que una asignación aleatoria.

Si el ARI es cercano a 1, significa que los grupos formados por K-Means son casi iguales a las categorías reales de las flores. Si el valor es cercano a 0, significa que los grupos fueron hechos casi al azar. Si es negativo, significa que los grupos están peor organizados de lo que estarían por azar.


##### La inercia

Cuanto menor sea la inercia, mejor será el agrupamiento en términos de cohesión interna, ya que los puntos estarán más cerca de sus centroides.

No se puede usar sola para evaluar el modelo porque siempre disminuirá a medida que aumente el número de clusters, incluso si no es un agrupamiento significativo.

Elbow Method (Método del Codo): Se usa la inercia para encontrar el número óptimo de clusters trazando la inercia frente a la cantidad de clusters. Se busca el "codo" de la gráfica, donde la reducción de la inercia se vuelve menos significativa.

#### Coeficiente de Silueta

Este coeficiente permite evaluar la compacidad y separación de los clusters.\
Valores cercanos a 1 indican alta cohesión y alta separación, valores cercanos a 0 indican solapamiento y valores negativos sugieren que los puntos están mal agrupados.


```python
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score, silhouette_score

# Cargar el dataset Iris
iris = load_iris()
X = iris.data  # Características (atributos)
y = iris.target  # Etiquetas (no se usarán en el clustering)

# Estandarizar las características para que tengan media = 0 y varianza = 1
# Esto es importante para K-means, ya que es sensible a la escala de los datos.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Método del Codo para determinar el número óptimo de clusters
inertias = []

#Metodo de la silueta
silhouette_scores = []

# Definir el número de clusters (k). En este caso, sabemos que son 3 clases.
k_values = range(2, 11)  # Probar valores de k entre 1 y 10

for k in k_values:
    # Crear el modelo K-means
    # Con n_init=X, K-means ejecuta el algoritmo X veces, cada vez con una inicialización diferente de los centroides.
    # Después de cada ejecución, el algoritmo calcula la inercia y selecciona la ejecución que tuvo el menor valor y que generar
    # el mejor resultado.
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    
    # Entrenar el modelo con los datos estandarizados
    # Observa que los datos no están etiquetados
    kmeans.fit(X_scaled)
    
    # Obtener la inercia. Es una medida de lo compactos que son los clusteres
    # Un valor bajo de inercia indica que los puntos están muy cerca de sus centroides, lo que sugiere clusters bien definidos
    inertias.append(kmeans.inertia_)
    
    # Obtenemos el valor de la silueta entre -1 y 1. Mide qué tan bien se ha hecho el agrupamiento en relación con otros posibles agrupamientos. 
    # Un valor cercano a 1 indica que los puntos están bien agrupados, mientras que valores cercanos a -1 indican que los puntos están mal asignados.
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))
    
    # Obtener el número de iteraciones 
    n_iter = kmeans.n_iter_
    print("---------------------------------------------------------")
    print("Numero de clusters: ",k,"Número de iteraciones:", n_iter)
    print()
    # Obtener las etiquetas de los clusters
    labels = kmeans.labels_
    print("Clasificación generadas:", labels)
    print()
    # Comparar con las etiquetas reales (solo para análisis, no se usan en el clustering)
    print("Etiquetas reales:", y)
    print()
    
    # Obtener los centroides
    centroids = kmeans.cluster_centers_
    print("Centroides de los clusters:", centroids)
    print()
    
    # Comparamos con las etiquetas reales usando el Adjusted Rand Index 
    # Un valor cercano a 1 indica una coincidencia casi perfecta entre los clusters y las etiquetas reales.
    ari = adjusted_rand_score(y, labels)
    print("Adjusted Rand Index (comparación con etiquetas reales):", ari)
    print()
    
    
      
    # Visualizar los clusters formados
    # Vamos a graficar solo las dos primeras características para simplificar la visualización
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Centroides
    plt.title('Clusters generados por K-means')
    plt.xlabel('Característica 1 (estandarizada)')
    plt.ylabel('Característica 2 (estandarizada)')
    plt.show()
    
# Fin del for



# Gráficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertias, marker='o', linestyle='--')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo para Selección de k')
plt.xticks(k_values)
plt.grid()
plt.show()

# Graficar las puntuaciones de silueta para los diferentes valores de K
plt.plot(k_values, silhouette_scores, marker='o')
plt.title('Puntuación de Silueta para Diferentes Valores de K')
plt.xlabel('Número de Clústeres (K)')
plt.ylabel('Puntuación de Silueta')
plt.xticks(k_values)  # Etiquetas en el eje x para K
plt.grid(True)
plt.show()
```
