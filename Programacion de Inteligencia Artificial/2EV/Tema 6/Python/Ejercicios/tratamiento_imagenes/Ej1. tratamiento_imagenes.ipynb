## Tratamiento de imágenes en OpenCV

OpenCV es una biblioteca de código abierto altamente optimizada para la visión por computadora y el procesamiento de imágenes. Su uso se ha popularizado debido a la eficacia y la variedad de operaciones que permite realizar en imágenes, siendo fundamental para aplicaciones que van desde la seguridad hasta la automatización y el entretenimiento interactivo.

### Estructura de una Imagen

En OpenCV, las imágenes se tratan como matrices o arreglos de píxeles. Un píxel, la menor unidad de una imagen, contiene valores de intensidad o color que, en conjunto, forman la imagen completa. Estos valores varían dependiendo del tipo y la profundidad del formato de la imagen.

### Formatos de Color

OpenCV maneja varios formatos de color, pero los más comunes son BGR, RGB y Escala de Grises:

- **BGR**: Por defecto, OpenCV carga las imágenes en formato BGR (Azul, Verde, Rojo). Este es el formato estándar usado internamente en la biblioteca para representar imágenes a color.

- **RGB**: Comúnmente usado en aplicaciones que no son de OpenCV, como interfaces de usuario y procesamiento de imágenes en la web. RGB representa los colores rojo, verde y azul. Para visualizar imágenes correctamente en otras bibliotecas que utilizan el formato RGB, como `matplotlib`, es necesario convertir las imágenes de BGR a RGB.

- **Escala de Grises**: Este formato convierte una imagen a una matriz de intensidades de gris, donde cada píxel representa un tono de gris correspondiente a la luminancia original de los colores en la imagen. El uso de la escala de grises es común en aplicaciones de detección de movimiento, análisis de textura, y donde el color no aporta información adicional necesaria.


### Operaciones con las imagenes
Cuando tratamos con imágenes, podemos realizar operaciones de rotación, desplazamiento, escala, aumento de brillo, etc. Todo esto se puede luego utilizar para ampliar el dataset.


### Conversión entre Formatos

OpenCV proporciona funciones eficientes para convertir imágenes entre estos diferentes formatos de color. Estas conversiones son esenciales para la preparación de imágenes para diversos algoritmos de procesamiento y análisis. Además, entender cómo manipular estos formatos es crucial para la implementación correcta de cualquier sistema de procesamiento de imágenes que use OpenCV.


```python
import cv2
import matplotlib.pyplot as plt

# Cargar una imagen en color
image_color = cv2.imread('../imagenes/cara.jpg', cv2.IMREAD_COLOR)

# Convertir de BGR a RGB (a matplotlib le gusta RGB)
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(image_rgb)
plt.show()
```

```python
# Cargar una imagen en escala de grises
image_gray = cv2.imread('../imagenes/cara.jpg', cv2.IMREAD_GRAYSCALE)

# Mostrar la imagen con un mapa de grises
plt.imshow(image_gray, cmap='gray')
plt.show()
```

```python
# rotación de imágenes:
import numpy as np

# Rotar una imagen 45 grados
height, width = image_color.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 45, 1)
rotated_image = cv2.warpAffine(image_color, rotation_matrix, (width, height))
cv2.imwrite('../imagenes/cara_rotada45.jpg',rotated_image)
# Convertir de BGR a RGB (a matplotlib le gusta RGB)
image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(image_rgb)
plt.show()
```

```python
# ajuste de brillo: 
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen original en formato BGR
image_color = cv2.imread('../imagenes/cara.jpg')

# Convertir de BGR a RGB
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Aumentar el brillo de la imagen
brighter_image = cv2.add(image_rgb, 50)

# Disminuir el brillo de la imagen
darker_image = cv2.add(image_rgb, -50)

# Configurar los subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(6, 18))

# Mostrar la imagen original
axes[0].imshow(image_rgb)
axes[0].set_title('Imagen Original')
axes[0].axis('off')  # Ocultar los ejes

# Mostrar la imagen más brillante
axes[1].imshow(brighter_image)
axes[1].set_title('Más Brillante')
axes[1].axis('off')  # Ocultar los ejes

# Mostrar la imagen más oscura
axes[2].imshow(darker_image)
axes[2].set_title('Más Oscura')
axes[2].axis('off')  # Ocultar los ejes

# Mostrar la figura
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen original
image = cv2.imread('../imagenes/pomeranian.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB

# Aplicar un flip horizontal a la imagen
image_flipped_horizontal = np.fliplr(image_rgb)

# Aplicar un flip vertical a la imagen
image_flipped_vertical = np.flipud(image_rgb)

# Configurar los subplots
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(15, 5))  # Ajustar el tamaño según sea necesario

# Mostrar la imagen original
axes[0].imshow(image_rgb)
axes[0].set_title('Imagen Original')
axes[0].axis('off')  # Ocultar los ejes

# Mostrar la imagen con flip horizontal
axes[1].imshow(image_flipped_horizontal)
axes[1].set_title('Flip Horizontal')
axes[1].axis('off')  # Ocultar los ejes

# Mostrar la imagen con flip vertical
axes[2].imshow(image_flipped_vertical)
axes[2].set_title('Flip Vertical')
axes[2].axis('off')  # Ocultar los ejes

# Mostrar la figura
plt.show()
```

```python
# desplazamiento: 
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen original en formato BGR
image_color = cv2.imread('../imagenes/pomeranian.png')

# Convertir de BGR a RGB para la visualización correcta en Matplotlib
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Obtener las dimensiones originales de la imagen
height, width = image_rgb.shape[:2]

# Definir la matriz de traslación para mover la imagen 200 píxeles hacia la derecha y 100 hacia abajo
translation_matrix = np.float32([[1, 0, 200], [0, 1, 100]])

# Aplicar la traslación a la imagen
translated_image = cv2.warpAffine(image_rgb, translation_matrix, (width, height))

# Configurar los subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Mostrar la imagen original
axes[0].imshow(image_rgb)
axes[0].set_title('Imagen Original')
axes[0].axis('off')  # Ocultar los ejes

# Mostrar la imagen trasladada
axes[1].imshow(translated_image)
axes[1].set_title('Imagen Trasladada')
axes[1].axis('off')  # Ocultar los ejes

# Mostrar la figura
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen original en formato BGR
image_color = cv2.imread('../imagenes/mario.jpg')

# Convertir de BGR a RGB para la visualización correcta en Matplotlib
image_rgb = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

# Obtener las dimensiones originales de la imagen
height, width = image_rgb.shape[:2]

# Calcular las nuevas dimensiones
new_width = int(width * 0.5)
new_height = int(height * 0.5)

# Redimensionar la imagen
resized_image = cv2.resize(image_rgb, (new_width, new_height), interpolation=cv2.INTER_AREA)

# Configurar los subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Mostrar la imagen original
axes[0].imshow(image_rgb)
axes[0].set_title('Imagen Original')

# Mostrar la imagen redimensionada
axes[1].imshow(resized_image)
axes[1].set_title('Imagen Escalada')

# Mostrar la figura
plt.show()
```
