# Visualización y análisis exploratorio
Ejercicio 1: Crea un gráfico de barras para visualizar el promedio de popularidad
(`track_popularity`) por género (`playlist_genre`)

```python
# prompt: Crea un gráfico de barras para visualizar el promedio de popularidad
# (`track_popularity`) por género (`playlist_genre`)

# Agrupar por género y calcular el promedio de popularidad
genre_popularity = spotify_data.groupby('playlist_genre')['track_popularity'].mean()

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_popularity.index, y=genre_popularity.values)
plt.xlabel('Género')
plt.ylabel('Promedio de Popularidad')
plt.title('Promedio de Popularidad por Género')
plt.xticks(rotation=90)
plt.show()
```

Ejercicio 3: Analiza la tendencia de la popularidad (`track_popularity`) en función de las fechas de
lanzamiento (`track_album_release_date`)

```python
# prompt: Analiza la tendencia de la popularidad (`track_popularity`) en función de las fechas de
# lanzamiento (`track_album_release_date`)

# Convertir la columna 'track_album_release_date' a tipo datetime si aún no lo está
spotify_data['track_album_release_date'] = pd.to_datetime(spotify_data['track_album_release_date'])

# Agrupar por año de lanzamiento y calcular la popularidad media
yearly_popularity = spotify_data.groupby(spotify_data['track_album_release_date'].dt.year)['track_popularity'].mean()
print(yearly_popularity)
# Crear el gráfico de línea
plt.figure(figsize=(12, 6))
plt.plot(yearly_popularity.index, yearly_popularity.values)
plt.xlabel('Lanzamiento')
plt.ylabel('Popularidad Media')
plt.title('Tendencia de la Popularidad a lo largo del tiempo')
plt.grid(True)
plt.show()
```
